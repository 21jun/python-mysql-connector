```python
def clean_text(text):
    text = text.lower()
    # replace special chars and keep spaces
    text = re.sub(r"[ ](?=[ ])|[^A-Za-z0-9 ]+", '', text)
    text = re.sub(' +', ' ', text)  # replace multiple spaces and keep one
    return text


def word_count(gameName):
    global games
    words = games[gameName].split()
    counts = Counter(words)
    return {k: v for k, v in sorted(counts.items(), key=lambda item: item[1], reverse=True)}
```

```python
db = DataBase()
```

```python
SQL = """
SELECT appid, gameName, videoId, likeCount, replyCount, text FROM yt_comment_steam where language='en';
"""
```

```python
comments = db.to_df(SQL)
```

```python
games = {}
gameNames = comments['gameName'].unique()

for gameName in gameNames:
    text = comments[comments['gameName'] == gameName]['text']
    text = text.tolist()
    text = [clean_text(t) for t in text]
    text = ' '.join(text)
    games[gameName] = text
```
