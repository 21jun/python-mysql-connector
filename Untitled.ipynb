```python
from database.pymysql_conn import DataBase
import pandas as pd
from matplotlib import pyplot as plt
```

```python
db = DataBase()
```

```python
SQL = """
SELECT A.appid, A.gameName, avg(viewCount), avg(likeCount), avg(commentCount), max(viewCount), COUNT(*), B.avg_player_count FROM(
SELECT * FROM yt.yt_video_info where title_cnt > 0 or tags_cnt > 0 or desc_cnt > 0 or comment_cnt > 0) A 
INNER JOIN  
(SELECT * FROM yt.games) B ON A.appid = B.appid group by A.appid order by B.avg_player_count desc;
"""
```

```python
df=db.to_df(SQL)
```
